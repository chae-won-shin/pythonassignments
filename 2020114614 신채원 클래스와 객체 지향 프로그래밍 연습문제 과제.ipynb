```python
#2번
class Counter:
    def __init__(self, number=0):
        if(number < 100 and number > -1):
            self.__number = number
            
    def reset(self):
        self.__number = 0
            
    def inc(self):
        self.__number += 1
        if self.__number >= 100:
            self.__number = 0
        return self.__number
    
    def dec(self):
        self.__number -= 1
        if self.__number <= -1:
            self.__number = 0
        return self.__number
    
    def __str__(self):
        return 'C('+str(self.__number)+')'
    
    def __add__(self, other):
        result = self.__number + other.__number
        if(result >= 100): result = 0
        return Counter(result)

    def __sub__(self, other):
        result = self.__number - other.__number
        if(result <= -1): result = 0
        return Counter(result)
    
c1 = Counter(10)
c1.inc()
print('c1 = ', c1)

print()

c2 = Counter()
c2.inc()
c2.inc()
c2.dec()
print('c2 = ', c2)
c2.reset()
print('c2 = ', c2)   

print()

c1 = Counter(10)
c2 = Counter(20)
c3 = c1 + c2
c4 = c1 - c2
print('c3 = ', c3)
print('c4 = ', c4)
```

```python
#3번
class BankAccount:
    def __init__(self, name, account_num, balance = 0):
        self.__name = name
        self.__account_num = account_num
        self.__balance = balance
    
    def __str__(self):
        return self.__name + '님의 계좌 '+self.__account_num+'의 잔고는 '\
                +str(self.__balance)+'원입니다.'
    
    def get_name(self):
        return self.__name
    
    def get_account_num(self):
        return self.__account_num
    
    def get_balance(self):
        return self.__balance
    
    def deposit(self, money):
        self.__balance += money
        print(str(money)+'가 입금되었습니다. 잔고는 '+str(self.__balance)+'원입니다.')
        return self.__balance
    
    def withdraw(self, money):
        if(money > self.__balance):
            print('계좌 잔고는 '+str(self.__balance)+'원으로 인출 요구 금액 '+str(money)+'보다 작습니다.')
        else:
            self.__balance -= money
            print(str(money)+'가 출금되었습니다. 잔고는 '+str(self.__balance)+'원입니다.')
            return self.__balance
        
account1 = BankAccount('홍길동', '1234-0001')
print(account1)
account1.deposit(2000)
print(account1)
account1.withdraw(500)
print(account1)
account1.withdraw(5000)
```

```python
#4번
class Student:
    def __init__(self, name, student_id, korean_quiz=0, math_quiz=0, science_quiz=0, total_score = 0):
        self.__name = name
        self.__student_id = student_id
        self.__korean_quiz = korean_quiz
        self.__math_quiz = math_quiz
        self.__science_quiz = science_quiz
        self.__total_score = self.get_total_score()
        
    def __str__(self):
        self.__total_score = self.get_total_score()
        avg = self.get_avg_score()
        return '이름: ' + self.__name + ',  학번: ' + self.__student_id + \
                '\n국어성적 : ' + str(self.__korean_quiz) + ',  수학성적 : '+ str(self.__math_quiz) + ',  과학성적: '+ str(self.__science_quiz) + \
                '\n합계 : '+ str(self.__total_score) + ',  평균: ' + str(avg)
        
    def get_name(self):
        return self.__name
    
    def get_student_id(self):
        return self.__student_id
    
    def get_korean_quiz(self):
        return self.__korean_quiz
    
    def get_math_quiz(self):
        return self.__math_quiz
    
    def get_science_quiz(self):
        return self.__science_quiz
    
    def set_korean_quiz(self, score):
        self.__korean_quiz = score
    
    def set_math_quiz(self, score):
        self.__math_quiz = score
        
    def set_science_quiz(self, score):
        self.__science_quiz = score
        
    def get_total_score(self):
        self.__total_score = self.__korean_quiz + self.__math_quiz + self.__science_quiz
        return self.__total_score
    
    def get_avg_score(self): 
        return self.__total_score / 3
            
name = input('학생의 이름을 입력하세요 : ')
student_id = input('학생의 학번을 입력하세요 : ')
student = Student(name, student_id)

korean_quiz = int(input('학생의 국어 성적을 입력하세요 : '))
math_quiz = int(input('학생의 수학 성적을 입력하세요 : '))
science_quiz = int(input('학생의 과학 성적을 입력하세요 : '))
student.set_korean_quiz(korean_quiz)
student.set_math_quiz(math_quiz)
student.set_science_quiz(science_quiz)
print(student)
```

```python
#5번
class TV:
    MIN_VOLUME = 0
    MAX_VOLUME = 20
    MIN_CHANNEL = 0
    MAX_CHANNEL = 200
    
    def __init__(self, volume = 5, channel = 0, is_on = False):
        self.__volume = volume
        self.__channel = channel
        if(self.__volume != 5 or self.__channel != 0):
            self.__is_on = True
        else: self.__is_on = is_on
        
    def __str__(self):
        if (self.__is_on == False):
            return 'TV가 꺼짐 상태입니다.'
        else:
            return '볼륨 = '+str(self.__volume)+',  채널 = '+str(self.__channel)
        
    def toggle_power(self):
        if(self.__is_on == False):
            self.__is_on = True
        else: 
            self.__is_on = False
            
    def get_channel(self):
        return self.__channel
            
    def set_channel(self, choice):
        if (choice<0 or choice > 201):
            print('채널오류')
        else :
            self.__channel = choice
            
    def get_volume(self):
        return self.__volume
    
    def set_volume(self, choice):
        if(choice < 0 or choice < 20):
            print('볼륨오류')
        else:
            self.__volume = choice
            
    def volume_up(self):
        if(self.__volume == TV.MAX_VOLUME):
            print('최대 볼륨입니다.')
        else: 
            self.__volume += 1
            
            
    def volume_down(self):
        if(self.__volume == TV.MIN_VOLUME):
            print('최소 볼륨입니다.')
        else: 
            self.__volume -= 1
        
    def channel_up(self):
        if(self.__channel == TV.MAX_CHANNEL):
            self.__channel = 0
        else:
            self.__channel += 1
            
    def channel_down(self):
        if(self.__channel == TV.MIN_CHANNEL):
            self.__channel = 200
        else:
            self.__channel -= 1
            
my_tv = TV()
print(my_tv)
my_tv.toggle_power()
print(my_tv)
my_tv.set_channel(200)
print(my_tv)
my_tv.volume_up()
my_tv.channel_up()
print(my_tv)
```

```python
#6번
class Rectangle:
    def __init__(self, x, y, width, height):
        self.__x = x
        self.__y = y
        self.__width = width
        self.__height = height
        
    def __str__(self):
        area = self.area()
        return 'Rectangle : (x = ' +str(self.__x)+ ', y = ' + str(self.__y)+ ', h = '+str(self.__height)+'), 면적: '+ str(area)
    
    def set_x(self, x):
        self.__x = x
    
    def set_y(self, y):
        self.__y = y
        
    def set_width(self, width):
        self.__width = width
        
    def set_height(self, height):
        self.__height = height
        
    def get_x(self):
        return self.__x
    
    def get_y(self):
        return self.__y
    
    def get_width(self):
        return self.__width
    
    def get_height(self):
        return self.__height
    
    def area(self):
        return (self.__width)*(self.__height)
    
    def overlaps(self, rec):
        if (self.__x <= rec.__x <= self.__x + self.__width) \
        or (self.__x <= (rec.__x + rec.__width) <= self.__x + self.__width) :
            if(self.__y - self.__height <= rec.__y and rec.__y <= self.__y) \
            or (self.__y - self.__height <= rec.__y - rec.__height <= self.__y):
                return True
            
        else: return False
        
    def contains(self, rec):
        if(self.__x <= rec.__x) and (rec.__y <= self.__y):
            if(rec.__width < self.__width) and (rec.__height < self.__height):
                return True
        else: return False
    
r1 = Rectangle(0,0,100,100)
r2 = Rectangle(0, -10, 10, 10)
r3 = Rectangle(-100, 0, 120, 100)

print('r1 = ', r1)
print('r2 = ', r2)
print('r3 = ', r3)

print('r1 contains r2 : ', r1.contains(r2))
print('r1 contains r3 : ', r1.contains(r3))
print('r1 overlaps r2 : ', r1.overlaps(r2))
print('r1 overlaps r3 : ', r1.overlaps(r3))
```

```python
#7번:
class Circle:
    PI = 3.14
    def __init__(self, x, y, radius):
        self.__x = x
        self.__y = y
        self.__radius = radius
      
    def __str__(self):
        area = self.area()
        return 'Circle : (x = '+str(self.__x)+', y = '+str(self.__y)+', r = '+str(self.__radius)+'), 면적: '+ str(area)
    
    def set_x(self, x):
        self.__x = x
        
    def get_x(self):
        return self.__x
    
    def set_y(self, y):
        self.__y = y
        
    def get_y(self):
        return self.__y
    
    def set_radius(self, r):
        self.__radius = r
        
    def get_radius(self):
        return self.__radius
    
    def area(self):
        return (self.__radius**2)*Circle.PI
    
    def overlaps(self, c):
        distance = ((c.__x - self.__x)**2 + (c.__y - self.__y)**2)**0.5
        if distance <= (self.__radius + c.__radius):
            return True
        else: return False
        
    def contains(self, c):
        distance = ((c.__x - self.__x)**2 + (c.__y - self.__y)**2)**0.5
        if distance <= abs(self.__radius - c.__radius):
            return True
        else: return False
        
    
c1 = Circle(0, 0, 100)
c2 = Circle(0, -10, 10)
c3 = Circle(-100, 0, 120)

print('c1 = ', c1)
print('c2 = ', c2)
print('c3 = ', c3)

print('c1 contains c2 : ', c1.contains(c2))
print('c1 contains c3 : ', c1.contains(c3))
print('c1 overlaps c2 : ', c1.overlaps(c2))
print('c1 overlaps c3 : ', c1.overlaps(c3))

#면적 값은 교재 572p에 나온 예시답안과 동일합니다.
        
```
